# K-means Clustering

Motivation:

Given a set of observation $(x_1, x_2, \dots, x_m)$, where each observation is a $d$-dimensional real vector, $k$-means clustering aims to partition the $n$ observations into $k$ $(\leq n)$ sets $\mathbf{S} = \{ S_1, S_2, \dots, S_k\}$ so as to minimize the within-cluster sum of squares (WCSS), that is, the variance.

The algorithm of $k$-means clustering works as follows:

1) Initialize cluster centroids $\mu_1$, $\mu_2$, $\dots$, $\mu_k$ $\in \mathbb{R}^n$

2) Repeat until convergence: 
- For every $i$, set 

$$c^{(i)} := \argmin_j ||x^{(i)} - \mu_j||^2$$

- For each $j$, set

$$\mu_j := \frac{\sum_{i=1}^{m} 1\{c^{(i)} = j\} x^{(i)}}{\sum_{i=1}^{m} 1\{c^{(i)} = j\}}$$

where $c^{(i)}$ is a label correspondent to the number of the cluster and $1\{c^{(i)} = j\}$ is 0 if $c^{(i)} \neq j$ and 1 if $c^{(i)} = j$.

```python
import pandas as pd
df = pd.read_excel('K_means.xlsx')
print(df.head())
```

```python
import seaborn as sns
sns.regplot(x=df['X'], y=df['Y'], fit_reg = False)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=0)
```

```python
model = kmeans.fit(df)
predicted_values = kmeans.predict(df)
```

```python
# Looking again at the input data

from matplotlib import pyplot as plt
plt.scatter(df['X'], df['Y'])
```

```python
# Looking at the output
plt.scatter(df['X'], df['Y'], c=predicted_values, s=50, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=200, c='black')
```
